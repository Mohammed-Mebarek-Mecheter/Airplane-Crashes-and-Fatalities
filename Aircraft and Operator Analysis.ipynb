```python
# Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the Dataset
df = pd.read_csv("cleaned_dataset.csv")
```

## Aircraft and Operator Analysis

### Common Aircraft Types

```python
# Group by aircraft type and count incidents
aircraft_types_incidents = df['AC Type'].value_counts().reset_index(name='Incident Count')
# Get the top aircraft types with the highest incidents
top_aircraft_types = aircraft_types_incidents.head(10)

# Display the top aircraft types
print("Top 10 Most Common Aircraft Types Involved in Incidents:")
print(top_aircraft_types)
```

```python
# Plotting the top aircraft types with the highest incidents
plt.figure(figsize=(12, 8))
sns.barplot(x='Incident Count', y='AC Type', data=top_aircraft_types, palette='rocket')
plt.title('Top 10 Most Common Aircraft Types Involved in Incidents (1908-2023)')
plt.xlabel('Number of Incidents')
plt.ylabel('Aircraft Type')
plt.show()
```

### Operator Patterns

```python
# Group by operator and count incidents
operator_incidents = df['Operator'].value_counts().reset_index(name='Incident Count')
# Get the top and bottom operators with the highest and lowest incidents
top_operators = operator_incidents.head(10)
bottom_operators = operator_incidents.tail(10)

# Display the top operators 
print("Top 10 Operators with the Highest Number of Airplane Incidents:")
print(top_operators)
```

```python
# bottom operators
print("\nBottom 10 Operators with the Lowest Number of Airplane Incidents:")
print(bottom_operators)
```

```python
# Plotting the top operators with the highest incidents
plt.figure(figsize=(12, 8))
sns.barplot(x='Incident Count', y='Operator', data=top_operators, palette='cividis')
plt.title('Top 10 Operators with the Highest Number of Airplane Incidents (1908-2023)')
plt.xlabel('Number of Incidents')
plt.ylabel('Operator')
plt.show()

# Plotting the bottom operators with the lowest incidents
plt.figure(figsize=(12, 8))
sns.barplot(x='Incident Count', y='Operator', data=bottom_operators, palette='viridis')
plt.title('Bottom 10 Operators with the Lowest Number of Airplane Incidents (1908-2023)')
plt.xlabel('Number of Incidents')
plt.ylabel('Operator')
plt.show()
```

### Operator Region Trends

```python
# Extract region or country information from 'Operator' and create a new column 'Operator Region'
df['Operator Region'] = df['Operator'].apply(lambda x: x.split('(')[-1].strip(')') if '(' in x else None)

# Group by operator region and count incidents
operator_region_incidents = df['Operator Region'].value_counts().reset_index(name='Incident Count')

# Plotting the operator regions with the highest incidents
plt.figure(figsize=(12, 8))
sns.barplot(x='Incident Count', y='Operator Region', data=operator_region_incidents.head(10), palette='twilight')
plt.title('Top 10 Operator Regions with the Highest Number of Airplane Incidents (1908-2023)')
plt.xlabel('Number of Incidents')
plt.ylabel('Operator Region')
plt.show()
```
