```python
# Import Libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load the Dataset
df = pd.read_csv("cleaned_dataset.csv")
```

## Global Distribution

```python
# Group by location and count incidents
location_incidents = df['Location'].value_counts().reset_index(name='Incident Count')
# Get the location with the highest incidents
highest_incident_location = location_incidents.iloc[0]['Location']
highest_incident_count = location_incidents.iloc[0]['Incident Count']

print(f"The location with the highest number of airplane incidents is {highest_incident_location} with {highest_incident_count} incidents.")
```

```python
# Plotting the top locations with the highest incidents
plt.figure(figsize=(14, 8))
sns.barplot(x='Incident Count', y='Location', data=location_incidents.head(10), palette='viridis')
plt.title('Top 10 Locations with the Highest Number of Airplane Incidents (1908-2023)')
plt.xlabel('Number of Incidents')
plt.ylabel('Location')
plt.show()
```

## Location-specific Factors

```python
# Extract city information from 'Location' and create a new column 'City'
df['City'] = df['Location'].apply(lambda x: x.split(', ')[0] if ', ' in x else None)
# Group by city and count incidents
city_incidents = df['City'].value_counts().reset_index(name='Incident Count')
# Get the top cities with the highest incidents
top_cities = city_incidents.head(10)

# Display the top cities
print("Top 10 Cities with the Highest Number of Airplane Incidents:")
print(top_cities)
```

```python
# Plotting the top cities with the highest incidents
plt.figure(figsize=(14, 8))
sns.barplot(x='Incident Count', y='City', data=top_cities, palette='magma')
plt.title('Top 10 Cities with the Highest Number of Airplane Incidents (1908-2023)')
plt.xlabel('Number of Incidents')
plt.ylabel('City')
plt.show()
```
